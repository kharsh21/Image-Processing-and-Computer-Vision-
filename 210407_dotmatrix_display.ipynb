```python
def circle(img1,x,y):
    img2=img1.copy()
    for m in range(x-25,x+26):
        for n in range(y-25,y+26):
            
            if ((m-x)**2)+((n-y)**2)<=625:
                    img2[m][n]=255
    
    return img2
```
